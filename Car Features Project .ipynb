# Quantile for the all column

```python
Q1 = df.quantile(numeric_only=True,axis=0,q=0.25)
Q1
```

```python
Q2 = df.quantile(numeric_only=True,axis=0,q=0.50)
Q2
```

```python
Q3 = df.quantile(numeric_only=True,axis=0,q=0.75)
Q3
```

```python
IQR = Q3-Q1
IQR
df.shape
```

# Removing outlier based on the interquantile tange condition 

```python
df = df[~((df < (Q1-1.5*IQR)) | (df > (Q3+1.5*IQR))).any(axis=1)]
```
